```python
def get_tags(url):
    html_doc = ''
    with urlopen(url) as response:
        for line in response:
            line = line.decode('utf-8')
            html_doc = html_doc + line.replace('\n','')
    soup = BeautifulSoup(html_doc, 'html.parser')
    tags_list = soup.find('ul',{'class':'tags-list'}).find_all('li')
    return tags_list
```

```python
def add_tags(url):
    global all_pages
    html_doc = ''
    with urlopen(url) as response:
        for line in response:
            line = line.decode('utf-8')
            html_doc = html_doc + line.replace('\n','')
    soup = BeautifulSoup(html_doc, 'html.parser')
    items = soup.find_all('div',{'class':'eagle-item'})
    urls = []
    for item in items:
        url = base_url + item.a['href']
        all_pages.append(url)
        urls.append(url)
    return urls
```

```python
url = 'https://www.bbc.com/punjabi/india-46959990'
d = collections.defaultdict(lambda: [])
cur = 0
all_pages = [url]
patience = 0
prev = 0
while cur < len(all_pages):
    tags_list = get_tags(all_pages[cur])
    for l in tags_list:
        url = base_url + l.a['href']
        d[l.a.text].extend(add_tags(url))
    cur = cur + 1
    if len(set(all_pages)) == prev:
        patience = patience + 1
    else:
        prev = len(set(all_pages))
        patience = 0
    if patience > 20:
        break
```

```python
def clean_text(text):
    text = text.replace('ਇਸ ਨਾਲ ਸਾਂਝਾ ਕਰੋ','')
    text = text.replace('FACEBOOK','')
    text = text.replace('INSTAGRAM','')
    text = text.replace('TWITTER','')
    text = text.replace('YouTube','')
    text = text.replace('Facebook','')
    text = text.replace('Messenger','')
    text = text.replace('Twitter','')
    text = text.replace('WhatsApp','')
    return text
```

```python
def get_text(url):
    html_doc = ''
    with urlopen(url) as response:
        for line in response:
            line = line.decode('utf-8')
            html_doc = html_doc + line.replace('\n','')
    soup = BeautifulSoup(html_doc, 'html.parser')
    text = soup.find('div',{'class':'story-body'}).text
    return clean_text(text)
```

```python
all_classes = []
for value in data_frame['tags'].values:
    all_classes.extend(value.split(','))
```

```python
problemTagsInId = []
for tags in data_frame['tags'].values:
    pTags = []
    for tag in tags.split(','):
        pTags.append(class_to_id[tag])
    problemTagsInId.append(pTags)
```

```python
one_hot_problem_tags = np.zeros((len(data_frame['tags'].values), nb_classes))
```

```python
for i,tags in enumerate(problemTagsInId):
    one_hot_problem_tags[i] = np.eye(nb_classes)[tags].sum(axis=0)
```
